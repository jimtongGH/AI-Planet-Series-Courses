# Pyplot Formatting

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

sma_data = pd.read_csv('data/Standard_Metropolitan_Areas_Data-data.csv')
```

## 1. Plotting consecutive plots  绘制连续图

```python
plt.plot(sma_data.work_force, sma_data.income, color='red')
plt.show()

plt.plot(sma_data.physicians, sma_data.income, color='blue')
plt.show()
```

```python
# Combining the 2 plots into 1         将两个图合并为一个
plt.plot(sma_data.work_force, sma_data.income, color='red')

plt.plot(sma_data.physicians, sma_data.income, color='blue')
plt.show()
```

## 2. Making the plots descriptive  使图更描述性

```python
# 1. Adding a Title
# 1. 添加标题

plt.scatter(sma_data.percent_senior, sma_data.crime_rate)
plt.title('Plot of Crime Rate vs. Percent Senior Citizens')
plt.show()
```

```python
# 2. Adding Labels
# 2. 添加标签

plt.scatter(sma_data.percent_senior, sma_data.crime_rate)

plt.title('Plot of Crime Rate vs. Percent Senior Citizens')

plt.xlabel('Percent Senior')
plt.ylabel('Crime Rate')
plt.show()
```

```python
# 3. Adding a legend
# 3. 添加图例

plt.plot(sma_data.work_force, sma_data.income, color='red', label='Work Force')
plt.plot(sma_data.physicians, sma_data.income, color='blue', label='Physicians')

plt.title('Income vs. Work Force and Physicians')
plt.legend() # Add legend to the plot

plt.show()
```

## 3. Formatting the plots  格式化图表

```python
# 1. Changing Colours
# 1. 更改颜色
plt.plot(sma_data.physicians, sma_data.income, color='blue', label='Physicians') # specify color by name
plt.plot(sma_data.work_force, sma_data.income, color='c', label='Work Force') # short color name (rgbcyk)
plt.plot(sma_data.hospital_beds, sma_data.income, color='0.75', label='Hospital Beds') # grayscale value (0-1)

plt.legend()

plt.show()
```

```python
# 2. Changing Line Styles
# 2. 更改线型
plt.plot(sma_data.work_force, sma_data.income, linestyle='dashed')
# plt.plot(sma_data.work_force, sma_data.income, linestyle='--') # same as above

## other line styles
# plt.plot(sma_data.work_force, sma_data.income, linestyle='solid') # solid line
# plt.plot(sma_data.work_force, sma_data.income, linestyle='-') # same as above

# plt.plot(sma_data.work_force, sma_data.income, linestyle='dashdot') # dash-dot line
# plt.plot(sma_data.work_force, sma_data.income, linestyle='-.') # same as above

# plt.plot(sma_data.work_force, sma_data.income, linestyle='dotted') # dotted line
# plt.plot(sma_data.work_force, sma_data.income, linestyle=':') # same as above

plt.xlabel('Work Force')
plt.ylabel('Income')

plt.show()
```

```python
# 3. Adding Markers
# 3. 添加标记

# plt.plot(sma_data.work_force, sma_data.income, marker='o', label='Work Force') # circle marker
plt.plot(sma_data.physicians, sma_data.income, marker='s', label='Physicians') # square marker
# plt.plot(sma_data.hospital_beds, sma_data.income, marker='^', label='Hospital Beds') # triangle marker
# plt.plot(sma_data.percent_senior, sma_data.income, marker='x', label='Percent Senior') # x marker

plt.xlabel('Physicians')
plt.ylabel('Income')

plt.show()
```

```python
# 4. Combining the 3 formatting options
# 4. 结合三种格式选项

plt.plot(sma_data.work_force, sma_data.income, color='red', linestyle='dashed', marker='o', label='Work Force')
# Shortcut above
plt.plot(sma_data.physicians, sma_data.income, '--sb', label='Physicians') # blue dashed line with square markers

# Scatter Plot with plot function 使用 plot 函数创建散点图
plt.plot(sma_data.hospital_beds, sma_data.income, 'g^', label='Hospital Beds') # green triangle markers

plt.xlabel('Work Force')
plt.ylabel('Income')

plt.legend()

plt.show()
```

## 4. Changing the size of the plots  更改图表大小

```python
# Changing the size of the plot
plt.figure(figsize=(10, 5)) # width, height in inches

plt.plot(sma_data.work_force, sma_data.income, color='red', linestyle='dashed', marker='o', label='Work Force')

plt.xlabel('Work Force')
plt.ylabel('Income')

plt.legend()

plt.show()
```

## 5. Multiple Plots in one figure  一个图形中包含多个图形
- The .subplot() method is used to add multiple plots in one figure. It takes three arguments: 使用 .subplot() 方法在 1 个图形中添加多个图表。它接受三个参数：
    - nrows: number of rows in the figure  nrows: 图形中的行数
    - ncols: number of columns in the figure  ncols: 图形中的列数
    - index: index of the plot  index: 图的索引

```python
# 1 row and 2 columns - Method 1
# 1 行 2 列 - 方法 1

plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
plt.plot(sma_data.work_force, sma_data.income, color='red', linestyle='dashed', marker='o', label='Work Force')
plt.title('Work Force vs. Income')

plt.subplot(1, 2, 2)
plt.plot(sma_data.physicians, sma_data.income, color='blue', linestyle='dashed', marker='o', label='Physicians')
plt.title('Physicians vs. Income')

plt.suptitle('Income') # Add a super title for the entire figure
plt.show()
```

```python
# 1 row and 2 columns - Method 2
# 1 行 2 列 - 方法 2

fig, (ax1, ax2) = plt.subplots(1, 2) # 1 row, 2 columns, first plot

fig.suptitle('Income')

ax1.plot(sma_data.work_force, sma_data.income, color='red', linestyle='dashed', marker='o', label='Work Force')
ax1.set_title('Work Force vs. Income')

ax2.plot(sma_data.physicians, sma_data.income, color='blue', linestyle='dashed', marker='o', label='Physicians')
ax2.set_title('Physicians vs. Income')

plt.show()
```

```python
# Overlapping Values  重叠值
# Simply put sharey=True inside the subplot function. The two subplots will now share the y axis values.
# 简单来说，在 subplot 函数中设置 sharey=True。现在这两个子图将共享 y 轴的值。

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True) # 1 row, 2 columns, first plot

fig.suptitle('Income')

ax1.plot(sma_data.work_force, sma_data.income, color='red', linestyle='dashed', marker='o', label='Work Force')
ax1.set_title('Work Force vs. Income')

ax2.plot(sma_data.physicians, sma_data.income, color='blue', linestyle='dashed', marker='o', label='Physicians')
ax2.set_title('Physicians vs. Income')

plt.show()
```

```python
# 2 rows and 1 column
# 2 行 1 列

fig, (ax1, ax2) = plt.subplots(2, 1) # 2 rows, 1 column
fig.suptitle('Income')
ax1.plot(sma_data.work_force, sma_data.income, color='red', linestyle='dashed', marker='o', label='Work Force')
ax1.set_title('Work Force vs. Income')

ax2.plot(sma_data.physicians, sma_data.income, color='blue', linestyle='dashed', marker='o', label='Physicians')
ax2.set_title('Physicians vs. Income')

plt.show()
```

```python
# 2 rows and 2 columns
# 2 行 2 列

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2) # 2 rows, 2 columns

fig.suptitle('Income')

ax1.plot(sma_data.work_force, sma_data.income, color='red', linestyle='dashed', marker='o', label='Work Force')
ax1.set_title('Work Force vs. Income')

ax2.plot(sma_data.physicians, sma_data.income, color='blue', linestyle='dashed', marker='o', label='Physicians')
ax2.set_title('Physicians vs. Income')

ax3.plot(sma_data.hospital_beds, sma_data.income, color='green', linestyle='dashed', marker='o', label='Hospital Beds')
ax3.set_title('Hospital Beds vs. Income')

ax4.plot(sma_data.percent_senior, sma_data.income, color='purple', linestyle='dashed', marker='o', label='Percent Senior')
ax4.set_title('Percent Senior vs. Income')

plt.show()
```
