## 1. Basic Plots I - Scatter Plot, Line Plot  基础图表 I - 散点图，折线图

### a. Scatter Plot 散点图
- A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for different numeric variables.  散点图（也称为散点图、散列图）使用点来表示不同数值变量的值。
- The position of each dot on the horizontal and vertical axis indicates values for an individual data point.  每个点在水平和垂直轴上的位置表示单个数据点的值。
- Scatter plots are used to observe relationships between variables.  散点图用于观察变量之间的关系。
### Applications of Scatter Plot  散点图的用途
- A scatter plot can be useful for identifying other patterns in data.  散点图有助于识别数据中的其他模式。
- We can divide data points into groups based on how closely sets of points cluster together.  我们可以根据点集聚集的紧密程度将数据点分组。
- Scatter plots can also show if there are any unexpected gaps in the data and if there are any outlier points. (Look at the 2 points away from rest of the data in the scatter plot. Those are outliers.)  散点图还可以显示数据中是否存在任何意外的空白区域，以及是否存在任何离群点。（看看散点图中远离其他数据的两个点。那是离群点。）
- This can be useful if we want to segment the data into different parts, like categorising users into different groups. 如果我们想将数据分割成不同的部分，比如将用户分类到不同的组中，这会很有用。

```python
# To create a scatter plot in Matplotlib we can use the .scatter() method:
# 要在 Matplotlib 中创建散点图，我们可以使用 .scatter() 方法：
plt.scatter(sma_data.crime_rate, sma_data.percent_senior)


# Is plt.show( ) always required?
# plt.show( ) 是否总是必需的？
# You might’ve observed the line plt.show( ) after the plt.scatter( ). Is it necessary to use?
# 您可能注意到在 plt.scatter( ) 后面有一行 plt.show( )，这是必须使用的吗？

# If Matplotlib is used in a terminal, scripts or specialized IDEs such as Spyder, Pycharm or VS Code, plt.show() is a must.
# 如果在使用 Matplotlib 的终端、脚本或专门的 IDE（如 Spyder、Pycharm 或 VS Code）时，plt.show()是必须的。
# If Matplotlib is used in an iPython shell or a notebook as Jupyter Notebook or Colab Notebook, plt.show() is usually unnecessary.
# 如果在使用 Matplotlib 的 iPython shell 或笔记本（如 Jupyter Notebook 或 Colab Notebook）时，plt.show()通常是不必要的。

plt.show()
```

### b. Line Plot 折线图
- A line chart is used to represent data over a continuous time span. It is generally used to show trend of a variable over time. Data values are plotted as points that are connected using line segments.  折线图用于表示跨越连续时间跨度内的数据。它通常用于展示变量随时间的变化趋势。数据值被绘制为点，这些点通过线段连接。

```python
# Creating a Line Plot (with 2 arguments)
# 创建线形图（带 2 个参数）
# plot() is a versatile command, and can take an arbitrary number of arguments.
# plot()是一个功能强大的命令，可以接受任意数量的参数。

plt.plot(sma_data.work_force, sma_data.income)
plt.show()
```

```python
# Creating a Line Plot (with a single argument)
# 创建线形图（单个参数）

# If you provide a list of n elements to the .plot() function, matplotlib will assume it is a sequence of y values, and automatically generates the x values for you as a range of n elements starting from 0.
# 如果你向.plot()函数提供一个包含 n 个元素的列表，matplotlib 会将其视为 y 值的序列，并自动为你生成 x 值，范围从 0 开始，包含 n 个元素。

# Since python ranges start with 0, the default list x has the same length as y. Hence the x data will be [0,1,2,3]. (Length same as y(4) but starts from 0 instead).
# 由于 Python 范围从 0 开始，默认列表 x 的长度与 y 相同。因此 x 数据将是[0,1,2,3]。（长度与 y(4)相同，但起始值为 0）。

plt.plot([1, 2, 3, 4])
plt.show()
```

## 2. Basic Plots II - Histogram, Bar Plot, Pie Chart  基础图表 II - 直方图，条形图，饼图

### a. Histogram 直方图
- A histogram is a graphical display of data using bars(rectangles) of different heights.  直方图是使用不同高度的条形（矩形）显示数据的图形。
- Parts of a Histogram  直方图的组成部分
    - The title: The title describes the information included in the histogram. 标题：标题描述了直方图中包含的信息。
    - X-axis: The X-axis are intervals that show the scale of values which the measurements fall under. These intervals are also called bins.  X 轴：X 轴是显示测量值所属的值尺度的间隔。这些间隔也称为箱子。
    - Y-axis: The Y-axis shows the number of times that the values occurred(frequency) for each interval on the X-axis.  Y 轴：Y 轴显示了 X 轴上每个间隔的值出现的次数（频率）。
    - The bars: The height of the bar shows the number of times the values occurred within the interval, while the width of the bar shows the covered interval.  条形：条形的高度显示了值在间隔内出现的次数，而条形的宽度显示了覆盖的间隔。

```python
# Matplotlib can be used to create histograms using the hist() method.
# 使用 Matplotlib 的 hist()方法可以创建直方图。

plt.hist(sma_data.percent_senior)
plt.show()
```

### b. Bar Plot 条形图
- Bar charts are one of the most common graphs used to show data associated with categorical variables.  条形图是用于展示与分类变量相关数据的最为常见的图表之一。

```python
# Creating a Vertical Bar Plot
# 创建垂直条形图
plt.bar(sma_data.region, sma_data.crime_rate)
plt.show()
```

```python
# Creating a Horizontal Bar Plot
# 创建水平条形图

plt.barh(sma_data.region, sma_data.crime_rate)
plt.show()
```

### c. Pie Chart 饼图
- A pie chart (or a circle chart) is a circular statistical graphic, which is divided into slices(wedges) to illustrate numerical proportion.  饼图（或圆形图）是一种圆形统计图形，它被分成若干份（扇形）以说明数值比例。
- Parameters of a pie chart  饼图的参数
    - x: The wedge sizes. x: 块的大小。
    - labels: A sequence of strings providing the labels for each wedge.  标签：提供每个扇区标签的字符串序列。
    - Colors: A sequence of colors through which the pie chart will cycle. If None, will use the colors in the currently active cycle.  颜色：饼图将循环使用的颜色序列。如果为 None，将使用当前活动循环中的颜色。
    - Autopct: string, used to label the wedges with their numeric value. The label will be placed inside the wedge. The format string will be fmt%pct.  Autopct：用于用数值标签扇区的字符串。标签将放置在扇区内部。格式字符串为 fmt%pct。

```python
# Pie chart can be made using the method pie().
# 饼图可以使用 pie() 方法创建。

firms = ['Firm A', 'Firm B', 'Firm C', 'Firm D', 'Firm E']
market_share = [15, 30, 45, 10, 5]

# explode: If you want to "explode" a slice, i.e. separate it from the pie, you can use the explode parameter.
Explode = (0.1, 0, 0, 0, 0)  # only "explode" the 1st slice

plt.pie(market_share, explode=Explode, labels=firms, autopct='%1.2f%%',)
plt.show()
```
